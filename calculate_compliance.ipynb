## read start and end date

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# read Tracking sheet
file_path = 'Tracking.xlsx'
sheet_name = 'Dates'


tracking = pd.read_excel(file_path, sheet_name=sheet_name)
tracking.set_index('email',inplace=True)
tracking['start_date'] = pd.to_datetime(tracking['start_date'])
tracking['end_date'] = pd.to_datetime(tracking['end_date'])

tracking
```

## calculating wearing hours for a specific patient using heart rate timeseries data

```python
# path for heart rate csv
input_path = 'heart_rate'
# user name
input_user = 'pps006'

# set max print lines
pd.set_option('display.max_rows', 100000)

# read start date and end date from tracking
user_tracking_row = tracking.loc[input_user]
start_date = user_tracking_row['start_date'].to_pydatetime()
end_date = user_tracking_row['end_date'].to_pydatetime()

# if end date is null or earlier than start date, please make sure the dates are correct
if (end_date is pd.NaT) or (start_date is pd.NaT):
    print("start date or end date is not a time")
elif(end_date <= start_date):
    print("start date later then end date")
else:
    daily_wearing_time = pd.read_csv(f'{input_path}/{input_user}.csv')
    daily_wearing_time['date'] = pd.to_datetime(daily_wearing_time['date'])
    
    daily_wearing_time = daily_wearing_time[(daily_wearing_time['date'] >= start_date) & (daily_wearing_time['date'] <= end_date)]
    
    #calculate wearing time for every day
    daily_wearing_time = daily_wearing_time.sort_values(by='date')
    daily_wearing_time = daily_wearing_time.groupby('date').size().reset_index(name='wearing_time')
    daily_wearing_time['wearing_time']/=60
    
    
    # create a data range index
    date_range = pd.date_range(start=start_date, end=end_date)
    daily_wearing_time.reset_index(drop=True)
    

    date_list = daily_wearing_time['date'].apply(lambda x: x.strftime('%Y-%m-%d')).values
    for date in date_range:
        if date.strftime('%Y-%m-%d') not in date_list:
            daily_wearing_time.loc[len(daily_wearing_time.index)] = [pd.to_datetime(date).date(),0] 
    daily_wearing_time['date'] = pd.to_datetime(daily_wearing_time['date'])
    daily_wearing_time = daily_wearing_time.sort_values(by = 'date')
    daily_wearing_time = daily_wearing_time.reset_index(drop=True)
    daily_wearing_time['date'] = daily_wearing_time['date'].apply(lambda x: x.strftime('%Y-%m-%d')).values

    print(daily_wearing_time)
```

## Plot wearing time by every single date

```python
plt.figure(figsize=(10, 6))
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(8))  # max 8 locator in x axis

plt.plot(daily_wearing_time['date'], daily_wearing_time['wearing_time'], marker='o', linestyle='-', color='skyblue', label='wearing_time',markersize=2)
# show standard 8 hours line
plt.axhline(8, color='red', linestyle='dashed', linewidth=2, label='8 hours')

# add title and label
plt.xlabel('Single day')
plt.ylabel('Wearing time')
plt.title('Wearing time / Date')
plt.xticks(rotation=45) 

plt.legend()
plt.show()
```

## Plot wearing time percentage by every week

```python
import matplotlib.ticker as mtick

weekly_wearing_time = daily_wearing_time
#if user wear more than 8 hours a day, counted as 8 hours
weekly_wearing_time['wearing_time'] = weekly_wearing_time['wearing_time'].apply(lambda x: min(x, 8))


# calculate every Monday for each week
def get_first_day_of_week(date):
    if isinstance(date, str):
        date = pd.to_datetime(date)
    return date - pd.DateOffset(days=date.weekday())

# aggregate wearing time by every weeks' Monday.
weekly_wearing_time['weekly_date'] = weekly_wearing_time['date'].apply(get_first_day_of_week)
weekly_wearing_time = weekly_wearing_time.groupby(weekly_wearing_time['weekly_date'])['wearing_time'].sum().reset_index()
# weekly_wearing_time['wearing_time'] = weekly_wearing_time['wearing_time'].apply(lambda x: "{:.2%}".format(x))
# print(weekly_wearing_time)



fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.plot(weekly_wearing_time['weekly_date'], weekly_wearing_time['wearing_time']/56*100, marker='o', linestyle='-', color='skyblue', label='wearing_time',markersize=2)


# add title and label
plt.xlabel('Week')
plt.ylabel('Weekly Compliance')
plt.title('Wearing time / Week')
plt.xticks(rotation=45) 

plt.legend()
plt.show()
```

## Calculate Compliance

```python
user_wearing_hours = daily_wearing_time['wearing_time'].sum()
print('Wearing hours:',user_wearing_hours)

total_wearing_hours = 8.0*((end_date-start_date).days+1)
print('Total wearing hours need:',total_wearing_hours)

compliance = round(100*user_wearing_hours/total_wearing_hours,2)
print(f'Compliance: {compliance}%')
```
